### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de cada tienda
ingreso1 = tienda["Precio"].sum()
ingreso2 = tienda2["Precio"].sum()
ingreso3 = tienda3["Precio"].sum()
ingreso4 = tienda4["Precio"].sum()

# Calcular el ingreso total general
ingreso_total = ingreso1 + ingreso2 + ingreso3 + ingreso4

# Imprimir los resultados
print(f"Ingreso Tienda : ${ingreso1:,.2f}")
print(f"Ingreso Tienda 2: ${ingreso2:,.2f}")
print(f"Ingreso Tienda 3: ${ingreso3:,.2f}")
print(f"Ingreso Tienda 4: ${ingreso4:,.2f}")
print(f"\nIngreso TOTAL de las 4 tiendas: ${ingreso_total:,.2f}")


```

# 2. Ventas por categoría

```python
tiendas = {
    "Tienda": pd.read_csv(url),
    "Tienda2": pd.read_csv(url2),
    "Tienda3": pd.read_csv(url3),
    "Tienda4": pd.read_csv(url4)
}

# Contar productos vendidos por Categoría del Producto en cada tienda
for nombre, tienda in tiendas.items():
    print(f"\n📊 Categorías más populares en {nombre}:")
    if "Categoría del Producto" in tienda.columns:
        conteo_categorias = tienda["Categoría del Producto"].value_counts()
        print(conteo_categorias)
    else:
        print("⚠️ La columna 'Categoría del Producto' no se encuentra en los datos.")
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificaciones
for nombre, tienda in tiendas.items():
    print(f"\n⭐ Promedio de calificaciones en {nombre}:")
    if "Calificación" in tienda.columns:
        tienda["Calificación"] = pd.to_numeric(tienda["Calificación"], errors="coerce")
        # convertir a numérico por si hay errores
        promedio = tienda["Calificación"].mean()
        print(f"Promedio: {promedio:.2f} / 5")
    else:
        print("⚠️ La columna 'Calificación' no se encuentra en los datos.")
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Analizar cada tienda
for nombre, url in urls.items():
    df = pd.read_csv(url)

    print(f"\n📦 Análisis de productos en {nombre}:")

    if "Producto" in df.columns:
        # Contar ventas por producto
        conteo_productos = df["Producto"].value_counts()

        # Mostrar el más y el menos vendido
        producto_mas_vendido = conteo_productos.idxmax()
        cantidad_mas_vendido = conteo_productos.max()

        producto_menos_vendido = conteo_productos.idxmin()
        cantidad_menos_vendido = conteo_productos.min()

        print(f"🔝 Producto más vendido: {producto_mas_vendido} ({cantidad_mas_vendido} ventas)")
        print(f"🔻 Producto menos vendido: {producto_menos_vendido} ({cantidad_menos_vendido} ventas)")

        # Visualización
        plt.figure(figsize=(10, 5))
        conteo_productos.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title(f"{nombre} - Top 10 productos más vendidos")
        plt.xlabel("Producto")
        plt.ylabel("Cantidad vendida")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    else:
        print("⚠️ La columna 'Producto' no se encuentra en los datos.")
```

# 5. Envío promedio por tienda

```python
urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}
# Calcular promedio del costo de envío por tienda
for nombre, url in urls.items():
    df = pd.read_csv(url)

    print(f"\n🚚 Costo de envío promedio en {nombre}:")

    if "Costo de envío" in df.columns:
        df["Costo de envío"] = pd.to_numeric(df["Costo de envío"], errors="coerce")
        promedio_envio = df["Costo de envío"].mean()
        print(f"Promedio: ${promedio_envio:,.2f}")
    else:
        print("⚠️ La columna 'Costo envío' no se encuentra en los datos.")
```

En esta tabla podemos ver la relación entre las tiendas que más facturan en relación a la calificación otorgada por los clientes:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos
data = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturación": [1150880400, 1116343500, 1098019600, 1038375700],
    "Calificación Promedio": [3.98, 4.04, 4.05, 4.00]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar el estilo visual
sns.set(style="whitegrid")

# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Facturación", y="Calificación Promedio", s=150, color="skyblue", edgecolor="black")

# Añadir etiquetas a cada punto
for i in range(len(df)):
    plt.text(df["Facturación"][i], df["Calificación Promedio"][i] + 0.005, df["Tienda"][i],
             ha='center', fontsize=10, fontweight='bold')

# Títulos y ejes
plt.title("Relación entre Facturación Total y Calificación Promedio por Tienda", fontsize=14)
plt.xlabel("Facturación Total (en miles de millones)", fontsize=12)
plt.ylabel("Calificación Promedio", fontsize=12)
plt.ylim(3.9, 4.1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
